### Using package: pytube is easier and simpler

```python
from pytube import YouTube

yt = YouTube('https://www.youtube.com/watch?v=DFCxiKXtKTI')
yt.streams.filter(only_audio=True, subtype='webm').order_by('bitrate').all()
```

### Downloading whole videos by a search

```python
from bs4 import BeautifulSoup as bs
import requests
from pytube import YouTube
base = "https://www.youtube.com/results?search_query="
qstring = "9gag"
r = requests.get(base+qstring)
print(base+qstring)
page = r.text
soup=bs(page,'html.parser')
soup
```

```python
vids = soup.select('.yt-uix-tile-link')

videolist = []
for vid in vids:
    tmp = 'https://www.youtube.com' + vid['href']
    videolist.append(tmp)
```

```python
videolist
```

```python
for item in videolist:
     if item.find('channel') == -1:
        print(item)
        yt = YouTube(item)
        video = yt.streams.filter(subtype='mp4', progressive=True).order_by('resolution').desc().first()
        print(video)
        video.download('./')
```
